```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../csv/run1479245320854.csv")
```

```python
data.head()
```

```python
car_2 = data[data.car_id == 2]
plt.plot(car_2.time, car_2.velocity, label = "1")

car_3 = data[data.car_id == 3]
plt.plot(car_3.time, car_3.velocity, label = "2")

car_4 = data[data.car_id == 4]
plt.plot(car_4.time, car_4.velocity, label = "3")

plt.legend(loc = "upper left")
```

```python
car_2 = data[data.car_id == 2]
plt.plot(car_2.time, car_2.distance, label = "1")

car_3 = data[data.car_id == 3]
plt.plot(car_3.time, car_3.distance, label = "2")

car_4 = data[data.car_id == 4]
plt.plot(car_4.time, car_4.distance, label = "3")

plt.legend(loc = "upper left")
```
