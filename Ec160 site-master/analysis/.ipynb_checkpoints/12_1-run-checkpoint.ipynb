```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../csv/round2-12_1.csv")
```

```python
data.head()
```

```python
for i in range(7):
    d = data[data.car_id == i]
    plt.plot(d.time, d.velocity, label = str(i))

plt.legend(loc = "upper left")
```

```python
for i in range(7):
    d = data[data.car_id == i]
    plt.plot(d.time, d.distance, label = str(i))

plt.legend(loc = "upper left")
```
